### Load Drug Use data from NHANES Web Page (API)

```python
!pip install nhanes_pytool_api
```

```python
from nhanes_data.nhanes_data_api import NHANESDataAPI
nhanes_api = NHANESDataAPI()

# Retrieve data from questionnaire - Drug Use - all cycle-years contained
cycle_years = nhanes_api.list_cycle_years()
drug_usage_df = nhanes_api.retrieve_data('questionnaire', cycle_years, 'Drug Use')


# subset the drug_usage_data to variables we're interested in
drug_df = drug_usage_df[['SEQN', 'year', 'DUQ120', 'DUQ200', 'DUQ250', 'DUQ290', 'DUQ330', 'DUQ370']]

# Based on the variable descriptions, for year cycles 1999-2000, 2001-2002, 2003-2004, the drug usage for specific drug types was missing,
# but the injection usage is included as DUQ120 (Ever used a needle to take street drugs)
# Merge DUQ120 for these 3 year cycles and DUQ370 (the injection usage after 2005) column
drug_df.loc[drug_df['year'].isin(["1999-2000", "2001-2002", "2003-2004"]), 'DUQ370'] = drug_df['DUQ120']

drug_df = drug_df[['SEQN', 'year', 'DUQ200', 'DUQ250', 'DUQ290', 'DUQ330', 'DUQ370']]
```

```python
df = demographic_subset.merge(
    questionnaire_subset, on=['SEQN', 'SDDSRVYR'], how='left'
).merge(
    mortality_subset, on=['SEQN', 'SDDSRVYR'], how='left'
)
df = df[df['SDDSRVYR'] != -1]

# Replace survey cycle values with the corresponding year range
survey_to_year = {
    1: '1999-2000',
    2: '2001-2002',
    3: '2003-2004',
    4: '2005-2006',
    5: '2007-2008',
    6: '2009-2010',
    7: '2011-2012',
    8: '2013-2014',
    9: '2015-2016',
    10: '2017-2018'
}
df['year'] = df['SDDSRVYR'].replace(survey_to_year)
df = df.drop(columns=['SDDSRVYR'])

df = df.merge(
    drug_df, on=['SEQN', 'year'], how='left'
)

df

cols = ['ID', 'Gender', 'Age', 'Race', 'Income', 'Education', 'Alcohol',
          'Cancer or malignancy', 'Congestive heart failure', 'Coronary heart disease', 'Angina/angina pectoris', 'Heart attack',
              'Mortality', 'Year', 'Marijuana or hashish', 'Cocaine', 'Heroin', 'Methamphetamine', 'Injection of illegal drug']
df.columns = cols
```

```python
df
```

### Map the number to the corresponding descriptions for each variable

```python
df_filled = df.copy()

# fill na for alcohol
df_filled['Alcohol'] = df_filled['Alcohol'].fillna(999)

# add a numeric year column
df_filled['Year'] = df_filled['Year'].str.split('-').str[1].astype(int)

df_filled
```

```python
# Convert numbers to words
num_to_gender = {
    1:'male',
    2:'female'
}
df_filled['Gender'] = df_filled['Gender'].replace(num_to_gender)

num_to_race = {
    1:'Mexican American',
    2:'Other Hispanic',
    3:'Non-Hispanic White',
    4:'Non-Hispanic Black',
    5:'Other',
    6:'Non-Hispanic Asian',
    7:'Other Race'
}
df_filled['Race'] = df_filled['Race'].replace(num_to_race).fillna("Don't know")

num_to_education = {
    1:'Less than 9th grade',
    2:'9-11th grade',
    3:'High school graduate/GED or equivalent',
    4:'Some college or AA degree',
    5:'College graduate or above',
    7:'Refused',
    9:"Don't know"
}
df_filled['Education'] = df_filled['Education'].replace(num_to_education).fillna("Don't know")

num_to_drug_disease = {
    1:'Yes',
    2:'No',
    7:'Refused',
    9:"Don't know"
}
drug_disease_cols = ['Cancer or malignancy', 'Congestive heart failure', 'Coronary heart disease', 'Angina/angina pectoris', 'Heart attack',
           'Marijuana or hashish', 'Cocaine', 'Heroin', 'Methamphetamine', 'Injection of illegal drug']
df_filled[drug_disease_cols] = df_filled[drug_disease_cols].replace(num_to_drug_disease).fillna("Don't know")

num_to_mortality = {
    0:'alive',
    1:'deceased'
}
df_filled['Mortality'] = df_filled['Mortality'].replace(num_to_mortality).fillna("Don't know")
```
